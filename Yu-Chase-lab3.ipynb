```python
social_graph = {
    "@bongolpoc":{"first_name":"Joselito",
                  "last_name":"Olpoc",
                  "following":[
                  ]
    },
    "@joaquin":  {"first_name":"Joaquin",
                  "last_name":"Gonzales",
                  "following":[
                      "@chums","@jobenilagan"
                  ]
    },
    "@chums" : {"first_name":"Matthew",
                "last_name":"Uy",
                "following":[
                    "@bongolpoc","@miketan","@rudyang","@joeilagan"
                ]
    },
    "@jobenilagan":{"first_name":"Joben",
                   "last_name":"Ilagan",
                   "following":[
                    "@eeebeee","@joeilagan","@chums","@joaquin"
                   ]
    },
    "@joeilagan":{"first_name":"Joe",
                  "last_name":"Ilagan",
                  "following":[
                    "@eeebeee","@jobenilagan","@chums"
                  ]
    },
    "@eeebeee":  {"first_name":"Elizabeth",
                  "last_name":"Ilagan",
                  "following":[
                    "@jobenilagan","@joeilagan"
                  ]
    },
}

def relationship_status(from_member,to_member):
    from_memberlist=social_graph[from_member]["following"]
    to_memberlist=social_graph[to_member]["following"]
    if to_member in from_memberlist and from_member in to_memberlist:
        return ("friends")
    elif to_member in from_memberlist:
        return ("follows")
    elif from_member in to_memberlist:
        return ("followed by")
    else:
        return ("")
        
print(relationship_status("@chums","@jobenilagan"))
print(relationship_status("@jobenilagan","@joeilagan"))
print(relationship_status("@bongolpoc","@chums"))
print(relationship_status("@bongolpoc","@jobenilagan"))
print(relationship_status("@bongolpoc","@chums"))
print(relationship_status("@chums","@joeilagan"))
print(relationship_status("@joeilagan","@chums"))
```

```python
board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

def get_winner (board):
    for i in range(len(board)):
        for j in range(len(board)):
            if board[0][j] == board[1][j] == board [2][j] == "X":
                return "X"
            elif board[0][j] == board[1][j] == board [2][j] == "O":
                return "O"
            elif board[i][0] == board[i][1] == board [i][2] == "X":
                return "X"
            elif board[i][0] == board[i][1] == board [i][2] == "O":
                return "O"
            elif i == j and board[0][0] == board[1][1] == board [2][2] == "X":
                return "X"
            elif i == j and board[0][0] == board[1][1] == board [2][2] == "O":
                return "O"
            elif i + j == 2 and board[0][2] == board[1][1] == board [2][0] == "X":
                return "X"
            elif i + j == 2 and board[0][2] == board[1][1] == board [2][0] == "O":
                return "O"
            else:
                return "none"
        
print (get_winner(board1))
print (get_winner(board2))
print (get_winner(board3))
print (get_winner(board4))
print (get_winner(board5))
print (get_winner(board6))
```
